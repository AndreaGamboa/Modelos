## CORRELACIONES VARIABLES

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
import statsmodels.api as sm
from statsmodels.graphics.api import abline_plot
from scipy import stats
from statsmodels import graphics
pd.set_option("display.max_rows", 10)
```

```python
# Importar conjunto de datos

variablestodas=pd.read_excel("TODAS LAS VARIABLES Y TOTAL.xlsx")
```

```python
#Eliminar valores faltantes del conjunto de datos

variablestodas=variablestodas.dropna()
```

### Correlación conjunto variables fotosintéticas

```python
foto=["Transpiracion", "Conductividad",
      "Asimilacion de CO2","Uso eficiente del agua"]
```

```python
#Gráficar en un mapa de calor las correlaciones

corMat=np.corrcoef([variablestodas[i] for i in foto])

fig, ax = plt.subplots(figsize=(5,5))
im = ax.imshow(corMat,cmap="RdBu")

# Mostrar las etiquetas de las variables

ax.set_xticks(np.arange(len(foto)))
ax.set_yticks(np.arange(len(foto)))


# Asignar los nombres a las filas y columnas

ax.set_xticklabels(foto[:5])
ax.set_yticklabels(foto[:5])

# Rotar los nombres
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Loop para asignar los valores de las correlaciones en el mapa de calor
for i in range(len(foto)):
    for j in range(len(foto)):
        text = ax.text(j, i, round(corMat[i, j],2),
                       ha="center", va="center", color="w")

plt.title("Matriz de correlaciones fotosíntesis", fontdict={"fontsize":15});
plt.savefig("Matriz de correlaciones fotosíntesis.jpg",dpi=110,bbox_inches="tight")
```
